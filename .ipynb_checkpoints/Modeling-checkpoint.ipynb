#### Conclusion:
The above build models were testwise just for predicting the change between month 29 and 30 (the latest two popularity levels in the dataset per search query) and having 29 months of popularity levels as features per search query. 
The goal is to use all available popularity levels per search query to build a model predicting the change in popularity for the next month (going over a defined treshold (1) or not (0)). 
This is achieved by the following:
- For each search query there are 232 months of popularity level data (from 01.01.2004 until 01.04.2023). 
- I take every month and the consecutive previous 29 months as one row. This is not working for the first 30 months, since there are no previous consecutive 30 months. They will be left out, so beginning on popularity level for month 31, there will be the calculation of change in popularity level between the second and last month of each row. With this technique, I can calculate the change between each month per search query beginning on month 31
- The model will get 29 months per row as features and should predict on this basis if the next month (which popularity level it does not know) will have a change in popularity level higher than a certain treshold compared to the last month (month 29). 

Why always 29 months and not having all popularity levels in one row per search query exept the last one:
- This is done under the assumption that popularity levels from the beginning (e.g. from 2006) will not have that much influence on the change of popularity level between e.g. March and April 2023, therefore only a subset of previous popularity levels (29) will be provided as features
- Splitting the dataset in junks of 29 months also prevents overfitting of the model on pattern which don't say anything(like the example above changes in popularity in 2006 will probably not have any influence on the popularity level change in March and April 2023)
- Creating more training data the model (not only one change between two months per search query, but changes in popularity level between nearly all available months for the search query

The above mentioned process is done below:

#### Define the function  to re-format the DataFrame to have one row per search query, now with all 232 months of popularity levels per search query

```python
def createHorizontalDataFrameAll(df):
    new_dataFrame = pd.DataFrame()
    for column in df.columns:
        column_own = df[['date', column]].copy()
        #column_own_last = column_own.iloc[-30:].reset_index(drop=True)
        column_list = [column]
        column_values = list(column_own[column])
        data_list_column = column_list + column_values
        df_column = pd.DataFrame([data_list_column])
        df_column = df_column.rename(columns={0: 'Query'})
        new_dataFrame = pd.concat([new_dataFrame, df_column], axis=0)
    
    #drop first row
    new_dataFrame = new_dataFrame.iloc[1:].reset_index(drop=True)
    return new_dataFrame
```

#### What was above done just for search query Nokia as an example for better understanding, will now be applied to all search queries in one DataFrame

```python
date_list = list(dataset['date'])
def getFinalDataset(df):
    final_df = pd.DataFrame()
    intermed_df_per_query = pd.DataFrame()
    for i in range(len(df)):
        query_name = df['Query'][i]
        one_query = df.iloc[i:i+1, ]
        #splitted_Nokia_df = pd.DataFrame()
        lst_columns = list(range(1,31))
        for j in range(1, len(one_query.columns) + 1):
            if j >=31:
                new_row = one_query.iloc[:, j-30:j]
                new_row.columns = lst_columns
                #display(new_row)
                intermed_df_per_query = pd.concat([intermed_df_per_query, new_row], axis=0)
                #display(splitted_Nokia_df)
            j = j + 1
        value_list = []
        for date in date_list[29:]:
            value = query_name + ' ' + date
            value_list.append(value)
        intermed_df_per_query.insert(0, "query_target_month", value_list)
        final_df = pd.concat([final_df, intermed_df_per_query], axis=0)
        intermed_df_per_query = pd.DataFrame()
        i = i + 1
    final_df = final_df.reset_index(drop=True)
    return final_df
```

#### Calculate how much better models are in comparison to randomly classify

```python
#Calculate how much standard deviations cumulative change is away from 0:
import math 
def calcStd(df):
    averageChangePos = df['change'][df['change'] > 0].mean()
    averageChangeNeg = df['change'][df['change'] <= 0].mean()
    sum_of_changes = df['change'].sum()
    sum_of_datapoints = len(df['change'])
    percentage_avg_change1 = abs(averageChangeNeg)/(abs(averageChangeNeg)+abs(averageChangePos))
    percentage_avg_change0 = 1 - percentage_avg_change1
    exp_value = 0
    variance = (percentage_avg_change0*(averageChangeNeg-exp_value)**2) + (percentage_avg_change1*(averageChangePos-exp_value)**2)
    stand_dev = math.sqrt(sum_of_datapoints)*math.sqrt(variance)
    stand_dev_final = sum_of_changes/stand_dev
    result = {'Standard_deviation': stand_dev_final}
    return result
```
