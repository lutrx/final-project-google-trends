```python
complete_dataset = pd.concat([data_1_42, data_43_177, data_178_184, data_185_202, data_203_271, data_272_401, 
                             data_402_447, data_448_468, data_469_498, data_499_539, data_540, data_541_578], axis=1)
```

```python
complete_dataset
```

```python
#list(complete_dataset.columns)
```

```python
def removeDuplicateColumns(df):
    df_new = df.loc[:,~df.columns.duplicated()]
    return df_new
```

```python
complete_dataset_corr = removeDuplicateColumns(complete_dataset)
```

```python
complete_dataset_corr
```

```python
complete_dataset_corr.to_csv('complete_dataset_cleaned.csv', index=False)
```
